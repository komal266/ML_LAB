```python
import pandas as pd

# Initialize an empty list to store company data
company_data = []

# Open and read the company data file
with open('C:/Users/Admin/Downloads/archive (1)/companies.csv') as file:
    for i, line in enumerate(file):
        # Split the line into columns based on comma separator
        data = line.strip().split(',')
        
        if i == 0:
            # Extract column names from the first line
            column_names = data
        else:
            # Check if the number of columns in this line matches the number of column names
            if len(data) != len(column_names):
                print(f"Error: Number of columns in line {i+1} doesn't match the number of column names.")
                continue
            
            # Append the company data to the list
            company_data.append(data)

# Create a DataFrame using the column names and company data
df = pd.DataFrame(company_data, columns=column_names)

# Display the first five entries in the DataFrame
print(df.head())
```
